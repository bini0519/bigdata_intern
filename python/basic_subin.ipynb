```python
from urllib.request import urlopen
import re
```

```python
#태풍정보
f = urlopen('https://www.weather.go.kr/plus/typ/report.jsp')
```

```python
type(f) #응답받음
```

```python
f.status 
#200은 정상적으로 응답받았다는 뜻
#404는 요청한 브라우저가 없다는 뜻
```

```python
f.getheader('Content-Type') #헤더 정보 알기
```

```python
# HTTP 헤더를 기반으로 인코딩 방식을 추출(명시돼 있지 않을 경우 utf-8을 사용)
encoding = f.info().get_content_charset(failobj="utf-8")
```

```python
print('encoding방식：', encoding)
```

```python
bytes_content = f.read()
```

```python
#임시 디코딩
scanned_text = bytes_content[:1024].decode('ascii', errors='replace')
```

```python
# 디코딩한 문자열에서 정규 표현식으로 charset 값을 추출
match = re.search(r'charset=["\']?([\w-]+)', scanned_text)
```

```python
if match:
    encoding = match.group(1)
else:
    # charset이 명시돼 있지 않으면 UTF-8을 사용
    encoding = 'utf-8'
```

```python
# 추출한 인코딩으로 다시 디코딩
text = bytes_content.decode(encoding)
# 응답 본문을 표준 출력에 출력
print(text)
```

```python
with open('typ.html', 'w', encoding=encoding) as wf:
    wf.write(text)
```
